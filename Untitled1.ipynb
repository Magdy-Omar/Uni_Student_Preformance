```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.pie(df['What is your favorite programming language?\n'].value_counts(), labels=df['What is your favorite programming language?\n'].value_counts().index, autopct='%1.2f%%');
plt.title("Favorite Programming Language");
```

```python

plt.figure(figsize=(9,6))

plt.scatter(data = df, x ='What is your total GPA', y='How many hours do you study per day outside of class?')



plt.xlabel("GPA")
plt.ylabel("Study Hours")
```

```python
plt.figure(figsize=(10, 6))
sns.regplot(x=df['What is your total GPA'], y=df['How many hours do you study per day outside of class?'], scatter_kws={'color': 'blue'}, line_kws={'color': 'red'})

plt.xlabel("GPA")
plt.ylabel("Study Hours")
plt.title("GPA vs Study Hours")

plt.show()

from scipy.stats import pearsonr
correlation, p_value = pearsonr(df['What is your total GPA'], df['How many hours do you study per day outside of class?'])
print(f"Pearson Correlation Coefficient: {correlation:.2f}")

if correlation > 0.5:
    print("Strong positive correlation: More study hours likely lead to a higher GPA.")
elif correlation > 0.2:
    print("Weak positive correlation: More study hours may have some effect on GPA.")
else:
    print("No significant correlation: Study hours and GPA do not have a strong relationship.")
```

```python
gpa_mean = df['What is your total GPA'].mean()
gpa_median = df['What is your total GPA'].median()
gpa_mode = df['What is your total GPA'].mode()[0]
gpa_std = df['What is your total GPA'].std()
study_hours_mean = df['How many hours do you study per day outside of class?'].mean()
study_hours_median = df['How many hours do you study per day outside of class?'].median()
study_hours_mode = df['How many hours do you study per day outside of class?'].mode()[0]
study_hours_std = df['How many hours do you study per day outside of class?'].std()
print(f"Mean GPA: {gpa_mean:.2f}")
print(f"Median GPA: {gpa_median:.2f}")
print(f"Mode GPA: {gpa_mode}")
print(f"Standard Deviation GPA: {gpa_std:.2f}")
print(f"Mean Study Hours: {study_hours_mean:.2f}")
print(f"Median Study Hours: {study_hours_median:.2f}")
print(f"Mode Study Hours: {study_hours_mode}")
print(f"Standard Deviation Study Hours: {study_hours_std:.2f}")
```
